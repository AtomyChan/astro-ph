```python
import os
from urllib.request import urlretrieve
import pandas as pd
import matplotlib.pyplot as plt

URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
def get_data(filename = 'FremontBridge.csv', url=URL, force_download = True):
    if force_download or not os.path.exists(url):
        urlretrieve(url, filename);
    data = pd.read_csv(filename, index_col = 'Date', parse_dates=True)
    data.columns = ['West', 'East']
    data.fillna(0, inplace=True)
    data['Total'] = data.eval('East + West')
    return data
```

```python
data = get_data()
```

```python
data.columns = ['West', 'East']
data.fillna(0, inplace=True)
data['Total'] = data.eval('East + West')
```

```python
import numpy as np
import seaborn; seaborn.set()  # plot styling
data.resample('W').sum().plot()
plt.ylabel('weekly trips');
```

```python
data.resample('D').sum().rolling(365).sum().plot()
plt.ylabel('rolling average trips');
```

```python
data.groupby(data.index.time).mean().plot()
```

```python
pivoted = data.pivot_table('Total',
                           index=data.index.time,
                           columns=data.index.date,
                           fill_value=0)
pivoted.plot(legend=False, alpha=0.01, c='k')
```

```python
pivoted = data.pivot_table(['East', 'West'],
                           index=data.index.date,
                           columns=data.index.hour,
                           fill_value=0)
pivoted
```

```python
X = pivoted.values
X
```

```python
from sklearn.decomposition import PCA
Xpca = PCA(0.85).fit_transform(X)
Xpca.shape
```

```python
total_trips = X.sum(1)
plt.scatter(Xpca[:, 0], Xpca[:, 1], c=total_trips,
            cmap='cubehelix')
plt.colorbar(label='total trips');
```
