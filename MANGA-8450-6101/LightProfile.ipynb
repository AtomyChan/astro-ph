```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.rcParams['figure.dpi'] = 100.0
plt.rcParams['figure.facecolor'] = 'w'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12.0
```

```python
profile1 = pd.read_csv('galaxy-3-r-fix.txt', sep='\s+')[1:]
profile2 = pd.read_csv('galaxy-3-r-free.txt', sep='\s+')[1:]
profile3 = pd.read_csv('galaxy-3-g-fix.txt', sep='\s+')[1:]
profile4 = pd.read_csv('galaxy-3-g-free.txt', sep='\s+')[1:]
```

```python
fig = plt.figure(figsize=(8,6))

SMA = np.array(profile1['SMA'][1:], dtype='float')
MAG = np.array(profile1['MAG'][1:], dtype='float')
lower_error = np.array(profile1['MAG_LERR'][1:], dtype='float')
upper_error = np.array(profile1['MAG_UERR'][1:], dtype='float')
asymmetric_error = [lower_error, upper_error]
plt.errorbar(SMA, MAG, yerr=asymmetric_error, fmt='*', markersize=5, capsize=3, c='r', label='Fixed e, PA')

SMA = np.array(profile2['SMA'][1:], dtype='float')
MAG = np.array(profile2['MAG'][1:], dtype='float')
lower_error = np.array(profile2['MAG_LERR'][1:], dtype='float')
upper_error = np.array(profile2['MAG_UERR'][1:], dtype='float')
asymmetric_error = [lower_error, upper_error]
plt.errorbar(SMA, MAG, yerr=asymmetric_error, fmt='o', markersize=5, capsize=3, c='r', label='Free e, PA')

SMA = np.array(profile3['SMA'][1:], dtype='float')
MAG = np.array(profile3['MAG'][1:], dtype='float')
lower_error = np.array(profile3['MAG_LERR'][1:], dtype='float')
upper_error = np.array(profile3['MAG_UERR'][1:], dtype='float')
asymmetric_error = [lower_error, upper_error]
plt.errorbar(SMA, MAG, yerr=asymmetric_error, fmt='*', markersize=5, capsize=3, c='g', label='Fix e, PA')

SMA = np.array(profile4['SMA'][1:], dtype='float')
MAG = np.array(profile4['MAG'][1:], dtype='float')
lower_error = np.array(profile4['MAG_LERR'][1:], dtype='float')
upper_error = np.array(profile4['MAG_UERR'][1:], dtype='float')
asymmetric_error = [lower_error, upper_error]
plt.errorbar(SMA, MAG, yerr=asymmetric_error, fmt='o', markersize=5, capsize=3, c='g', label='Free e, PA')


plt.gca().invert_yaxis()
plt.xlabel('Radius (pixel)')
plt.ylabel("MAG")
plt.legend()
plt.show()
```
