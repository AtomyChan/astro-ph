# Light Profile or `r` band

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.rcParams['figure.dpi'] = 100.0
plt.rcParams['figure.facecolor'] = 'w'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12.0
```

```python
for i in range(20):
    num = i+1
    filename = 'ra'+ str(num)+'.txt'
    data = pd.read_csv(filename, sep='\s+')
    data = data.drop(data.index[[0]])

    fig = plt.figure(figsize=(8,6))
    SMA = np.array(data['SMA'], dtype='float')
    MAG = np.array(data['MAG'], dtype='float')
    data['MAG_LERR'].values[[data['MAG_LERR']=='INDEF']]=np.nan
    data['MAG_UERR'].values[[data['MAG_UERR']=='INDEF']]=np.nan
    lower_error = np.array(data['MAG_LERR'], dtype='float')
    upper_error = np.array(data['MAG_UERR'], dtype='float')
    asymmetric_error = [lower_error, upper_error]
    plt.errorbar(SMA, MAG, yerr=asymmetric_error, fmt='*', markersize=5, capsize=3, c='r', label='Free e, PA')

    plt.gca().invert_yaxis()
    plt.xlabel('Radius (pixel)')
    plt.ylabel("MAG")
    plt.legend()
    plt.savefig('ra'+str(num)+'.png', dpi=300)
    plt.show()
```
